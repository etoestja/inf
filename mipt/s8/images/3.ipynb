```python
import numpy as np
import skimage
from skimage import data
from skimage.filters import rank
from skimage.morphology import disk
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
im = data.camera()
```

```python
nim = np.array(im, dtype=np.float64)
nim += np.random.rand(nim.shape[0], nim.shape[1]) * 100
```

```python
plt.imshow(nim, cmap='gray')
```

```python
selem = disk(5)
nnim = (nim - np.min(nim)) / (np.max(nim) - np.min(nim)) * 2 - 1
nnim = rank.mean(nnim, selem=selem)
```

```python
plt.imshow(nnim, cmap='gray')
```
